```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Charges': [100, 150, 120, 200, 180, 220]}

df = pd.DataFrame(data)

# Step 1: Group the DataFrame by the 'Category' column
grouped_data = df.groupby('Category')

# Step 2: Calculate the mean of the 'Charges' column for each group
mean_charges_per_group = grouped_data['Charges'].mean()

# Plotting the mean charges for each category
mean_charges_per_group.plot(kind='bar', color='skyblue')
plt.title('Mean Charges for Each Category')
plt.xlabel('Category')
plt.ylabel('Mean Charges')
plt.show()
```
